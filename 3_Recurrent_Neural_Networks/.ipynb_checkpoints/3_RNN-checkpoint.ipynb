## Part 3C. Pre-process Data for Keras 

```python
features_length = train_X.shape[1]
```

```python
# Standardize data 
scaler = StandardScaler()
scaler.fit(train_X)

scaled_X = {
    'train': scaler.transform(train_X),
    'validate': scaler.transform(validate_X),
    'test': scaler.transform(test_X),
    'score': scaler.transform(score_X),
}
```

```python
# check out shape of data & shape data SHOULD be for Keras processing purposes 

# Keras RNN cells expect inputs with 3 dimensions:
# (batch_size, condition_window, features_length)
# Each RNN cell takes a vector of features_length
# at each time step of condition_window.
# Internal RNN cell state is reset between different batch samples.

# Let's review current shapes of the input, and the desired shapes.

def expected_X_shape(actual_X_shape):
    return (int(actual_X_shape[0]/condition_window), condition_window, features_length)

def expected_y_shape(actual_y_shape):
    return (int(actual_y_shape[0]/prediction_window), prediction_window)


pd.DataFrame({
    'actual shape': [
        train_X.shape, 
        train_y.shape,
        validate_X.shape,
        validate_y.shape,
        test_X.shape,
        test_y.shape,
        score_X.shape,
    ],
    'expected shape': [
        expected_X_shape(train_X.shape), 
        expected_y_shape(train_y.shape), 
        expected_X_shape(validate_X.shape), 
        expected_y_shape(validate_y.shape), 
        expected_X_shape(test_X.shape), 
        expected_y_shape(test_y.shape), 
        expected_y_shape(score_X.shape), 
    ],
}, index=['train_X', 'train_y', 'validate_X', 'validate_y', 'test_X', 'test_y', 'score_X'])
```

```python
# change data into correct shape for Keras processing 

# Exepected shape can be achieved with a simple reshape call.
# Note that these are not Pandas data frames any more,
# but Numpy n-dimensional arrays.
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.reshape.html

def frame_to_ndarray(frame_X, frame_y):
    X_nd = frame_X.reshape(-1, condition_window, features_length).astype('float32')
    y_nd = frame_y['temperature'].values.reshape(-1, prediction_window).astype('float32')
    return X_nd, y_nd

train_X_nd, train_y_nd = frame_to_ndarray(scaled_X['train'], train_y)
validate_X_nd, validate_y_nd = frame_to_ndarray(scaled_X['validate'], validate_y)
test_X_nd, test_y_nd = frame_to_ndarray(scaled_X['test'], test_y)

# create def to transform ONLY score X (since we don't know actual values) 
def X_frame_to_ndarray(frame_X):
    X_nd = frame_X.reshape(-1, condition_window, features_length).astype('float32')
    return X_nd

score_X_nd = X_frame_to_ndarray(scaled_X['score'])
```

```python
# confirm shape of data 

pd.DataFrame({
    'actual shape': [
        train_X_nd.shape, 
        train_y_nd.shape,
        validate_X_nd.shape,
        validate_y_nd.shape,
        test_X_nd.shape,
        test_y_nd.shape,
        score_X_nd.shape,
    ],
    'expected shape': [
        expected_X_shape(train_X.shape), 
        expected_y_shape(train_y.shape), 
        expected_X_shape(validate_X.shape), 
        expected_y_shape(validate_y.shape), 
        expected_X_shape(test_X.shape), 
        expected_y_shape(test_y.shape), 
        expected_y_shape(score_X.shape), 
    ]},
    index=['train_X', 'train_y', 'validate_X', 'validate_y', 'test_X_nd', 'test_y_nd', 'score_X_nd']
)
```

## Part 4. Build Models

### Explanation of Architecture

Below are the changes I made to the original code that helped improve the performance of my RNN model. 

**Feature Engineering:**

I created two additional types of variables  time-based variables and location-based variables. 

For time-based variables, I used the time indexes of the datasets to separate out the attributes hour, day of year, month and quarter. These time-based attributes will allow the model to learn weather cycles that might be helpful in prediction. For example, midday is normally warmer than midnight, while June is normally warmer than December. This kind of information can help the model learn weather predictions better. Note that I didnt include variables such as year or day of week, as such variables would not help with understanding weather cycles. 

For location-based variables, I used the sample indexes of the datasets to separate out cities, which I then one-hot-encoded. Before including these additional variables, I noted that while the models performed well, it seemed like they were less able to follow localized temperature changes. As such, including location-based variables allowed the model to be more specific without needing to worry about overfitting. 

I also contemplated creating new weather-based variables, but am unsure if it would help the model. An idea I had was creating either lagged or moving average variables for temperature and humidity, but since the predictions from RNN model seemed smooth and dont react well to shocks, these additional variables seemed like it would further inhibit the model than help it. 

**Model Architecture:** 

In terms of model architecture, I changed the type of layers that were in the model. I first fitted the model with a LSTM layer, followed by a GRU layer, then a Dense layer to help predict output. I decided to use an LSTM layer first, as they remember longer sequences than GRUs, which I thought would be better for training purposes. Then, I used a GRU layer that utilizes its reset and update gates which I helped in predicting the input for the next sequence in the time series. Finally, I used a Dense layer as a decoder to make predictions. I tried adding more Dense layers and RNN layers into the model, but it seemed to decrease model accuracy than help it. 

Another important thing that I found useful was adding L1 and L2 regularizations to the model. Without the regularizations, I found that the training set had high R2 but the validation and testing sets had significantly lower R2s. I added an L1 regularization to the first LSTM layer as it helps with feature selection, which might be needed since I added so much additional variables to the dataset. I then added an L2 regularization to the second GRU layer, which helped with the overfitting issue. 

**Model Training:**

For model training, the main thing I changed was increasing the epoch size. I chose epoch size = 25 as it was large enough in size where it helped improved accuracy, but was also not redundant since increasing it after that point did not seem to help the model much. I changed my learning rate from 0.01 to 0.0075, and batch size from 100 to 80. At the same time, I also changed my Early Stopping and ReduceLROnPlateau parameters by increasing min_delta to 0.2 and changing mode of Early Stopping to min. I also changed the monitoring parameter from loss to val_loss since validation sets R2 was significantly lower, hence the parameter I was worried about. This occasionally helps automatically stop my model from training more epochs. This entire exercise was an endeavor of balancing training set R2 (which can always increase from more epochs) with the R2 of validation set and (later on) the R2 of test set.

```python
layers = [
    
    LSTM(input_shape=(condition_window, features_length), units=48,return_sequences=True, #reset_after=True,
        recurrent_activation='sigmoid', activation="tanh", kernel_regularizer=l1(0.1), 
        kernel_initializer=GlorotUniform(seed=42),recurrent_initializer=GlorotUniform(seed=43)),
    
    #BatchNormalization(),
    
    #Dense(64, activation='relu'), 
    
    BatchNormalization(),
    
    GRU(input_shape=(condition_window, features_length), units=48,return_sequences=False, reset_after=True,
        recurrent_activation='sigmoid', activation="tanh", kernel_regularizer=l2(0.2), 
        kernel_initializer=GlorotUniform(seed=42),recurrent_initializer=GlorotUniform(seed=43)),
    
    BatchNormalization(),
    
    #Dense(units=prediction_window,kernel_initializer=GlorotUniform(seed=36),activation='linear'), 
    
    Dense(units=prediction_window,kernel_initializer=GlorotUniform(seed=48),activation='linear')
]

model = Sequential(layers)
```

```python
model.summary()
```

```python
model.compile(
    optimizer=Adam(learning_rate=0.0075), # original 0.01
    loss='mean_squared_error',
    metrics=[
        'mean_absolute_error',
        mean_error,
        r2_score,
    ]
)
```

```python
# IMPORTANT: In order to pass the assignment, you need
# to improve the architecture and/or training procedure of this model.
# You CAN NOT submit the model as is. Please add plain English Python
# comments on how you modified the model and/or training procedure.

history = model.fit(
    train_X_nd,
    train_y_nd,
    validation_data=(validate_X_nd, validate_y_nd),
    epochs=25, # OG: 1
    batch_size=80, # OG: 100
    callbacks=[
        TerminateOnNaN(),
        ReduceLROnPlateau(
            monitor='val_loss',
            min_delta=0.2,
            patience=5,
            factor=0.1,
            cooldown=10,
            #mode='min',
            verbose=1
        ),
        EarlyStopping(
            monitor='val_loss',
            min_delta=0.2,
            patience=10,
            mode='min'
        )
    ]
)

metrics=['mean_absolute_error',
        mean_error,
        r2_score,
    ]
```

## Part 5. Check Models 

```python
# Note that Keras evaluates metrics in batches,
# which is okay for averaging-stable metrics like ME, MSE, and MAE,
# but R2 requires full dataset to evaluate precisely.

# This implies two things:
# 1) R2 as reported during training is not precise
# 2) when evaluating R2 after training, batch size must match dataset size

nd_array_by_segment = {
    'train': (train_X_nd, train_y_nd),
    'validate': (validate_X_nd, validate_y_nd), 
    'test': (test_X_nd, test_y_nd) # new 
}

metrics_by_segment = {
    segment: model.evaluate(
        X_nd, 
        y_nd, 
        verbose=0, 
        batch_size=X_nd.shape[0]
    )
    for segment, (X_nd, y_nd) in nd_array_by_segment.items()
}

rnn_metrics = pd.concat([
    pd.DataFrame(
        data=[metrics], 
        columns=['MSE', 'MAE', 'ME', 'R2'], 
        index=[segment]
    )
    for segment, metrics in metrics_by_segment.items()
])

rnn_metrics = pd.DataFrame([rnn_metrics.unstack()], index=['rnn'])\
    .swaplevel(axis=1, i=0, j=1)\
    .reindex(nd_array_by_segment.keys(), axis=1, level=0)\
    .round(4)

rnn_metrics
```
