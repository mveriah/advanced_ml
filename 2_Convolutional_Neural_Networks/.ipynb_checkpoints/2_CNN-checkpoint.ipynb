```python
# define function to load data 
def load_images(folder: str, limit: Optional[int] = None, preprocess: bool = True) -> np.ndarray:
    with zipfile.ZipFile('pnp_dataset.zip') as z:

        targets = sorted(p for p in z.namelist() if p.startswith(folder))
        if limit is not None:
            targets = targets[:limit]

        # Allocating an array for all images right away to avoid
        # storing images in a temporary Python list, which takes extra memory.
        images = np.empty((len(targets), 224, 224, 3), dtype=np.float16)

        for i, target in enumerate(targets):
            with z.open(target) as f:
                image_pixels = np.array(PIL.Image.open(f), dtype=np.float32)
                if preprocess:
                    image_pixels = preprocess_input(image_pixels)
                images[i] = image_pixels.astype(np.float16)
            
    return images
```

```python
# Load data in a smarter way 

# score_x -- keep as is 
score_x = load_images('pnp_dataset/score_x')

# train_y -- keep as is 
with zipfile.ZipFile('pnp_dataset.zip') as z:
    with z.open('pnp_dataset/train_y.npy') as f:
        train_y = np.load(f)

X_train, X_other, y_train, y_other = train_test_split(load_images('pnp_dataset/train_x', limit=80000), train_y[0:80000], test_size=0.2, random_state=42) 
X_validate, X_test, y_validate, y_test = train_test_split(X_other, y_other, test_size=0.5, random_state=42) 
```

# Part 2. Construction & Training of Model

### Part 2.1 Import & Set Up Model for Transfer Learning

```python
# Load model for transfer learning 

# https://arxiv.org/pdf/1801.04381.pdf
# https://keras.io/api/applications/mobilenet/#mobilenetv2-function
mobile_net = tf.keras.applications.MobileNetV2(
    include_top=False,
    input_shape=(224, 224, 3),
    weights="imagenet"
)
```

```python
# Freezes the parameters of the MobileNet layers, so they will not update during training.
# These parameters are initialized to a pre-trained snapshot using the ImagetNet dataset.
mobile_net.trainable = False
```

```python
# check out model
mobile_net.summary()
```

### Part 2.2 Set Up Model Infrastructure

```python
# this helps with managing computational resources 

with tf.device('/CPU:0'):
    # smallest items 
    X_validate = tf.constant(X_validate)
    y_validate = tf.constant(y_validate)
    X_test = tf.constant(X_test)
    y_test = tf.constant(y_test)
    
    # mid items
    score_x = tf.constant(score_x)
    train_y = tf.constant(train_y)
    
    # large items 
    X_train = tf.constant(X_train)
    y_train = tf.constant(y_train)
```

```python
model = tf.keras.models.Sequential([
    mobile_net,
    #tf.keras.layers.AvgPool2D(pool_size=(5, 5)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=512, activation="relu"), 
    tf.keras.layers.Dense(units=1, activation="sigmoid")
])
```

```python
model.compile(
        optimizer='sgd',
        loss='binary_crossentropy',
        metrics=[tf.keras.metrics.BinaryAccuracy(name="binary_accuracy",threshold=0.5)]
)
batch_size = 128 
```

### Part 2.3 Train Default Model 

```python
model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=batch_size, 
    verbose=1, 
    validation_data=(X_validate, y_validate)
)
```

```python
trained_X = model.predict(X_train)
validated_X = model.predict(X_validate)
tested_X = model.predict(X_test)

binary_training_accuracy = tf.keras.metrics.BinaryAccuracy()
binary_training_accuracy.update_state(y_train, trained_X)
default_model_train_accuracy = binary_training_accuracy.result().numpy()

binary_validated_accuracy = tf.keras.metrics.BinaryAccuracy()
binary_validated_accuracy.update_state(y_validate, validated_X)
default_model_validate_accuracy = binary_validated_accuracy.result().numpy()

binary_testing_accuracy = tf.keras.metrics.BinaryAccuracy()
binary_testing_accuracy.update_state(y_test, tested_X)
default_model_test_accuracy = binary_testing_accuracy.result().numpy()

print(default_model_train_accuracy,default_model_validate_accuracy,default_model_test_accuracy)
```

### Part 2.4 Hypertune Model 

```python
def build_model(encoder_width_1: int = 512, pooling_method: str = "no", pooling_size: int = 5, activation_type: str = "sigmoid") -> tf.keras.Model:
    
    if pooling_method == "no": 
        layers = [
            mobile_net, 
            tf.keras.layers.Flatten(),
            tf.keras.layers.Dense(units=encoder_width_1, activation="relu"),
            tf.keras.layers.Dense(units=1, activation="sigmoid")
        ]
    else: 
        if pooling_method == "avg": 
            pooling_layer = tf.keras.layers.AvgPool2D(pool_size=(pooling_size, pooling_size))
        elif pooling_method == "gbl": 
            pooling_layer = tf.keras.layers.GlobalAveragePooling2D()
        elif pooling_method == "max": 
            pooling_layer = tf.keras.layers.MaxPooling2D(pool_size=(pooling_size, pooling_size))
        layers = [
            mobile_net, 
            pooling_layer, 
            tf.keras.layers.Flatten(),
            tf.keras.layers.Dense(units=encoder_width_1, activation="relu"),
            tf.keras.layers.Dense(units=1, activation=activation_type)
        ]

    model = tf.keras.models.Sequential(layers)

    model.compile(
        optimizer='sgd',
        loss='binary_crossentropy',
        metrics=[tf.keras.metrics.BinaryAccuracy(name="binary_accuracy",threshold=0.5)]
    )
    
    return model
```

```python
# define train model 
def train_model(model: tf.keras.Model, epochs: int, verbose: int = 10, batch_size: int = 64) -> tf.keras.callbacks.History:
    history = model.fit(
        x=X_train, 
        y=y_train,
        epochs=epochs,
        batch_size=batch_size, #128
        shuffle=True,
        
        validation_data=(X_validate, y_validate), #(validate_noisy_X, validate_clean_X),
        #callbacks=[MlflowCallback()],
        verbose=verbose
    )
    
    return history
```

### Testing for Pooling Layer

```python
# Create parameters you want to test: 
# First round: average pooling layer, max layer, or no layer? 
r1_layer = ["no", "avg", "max", "glb"]
r1_train_accuracy = [] 
r1_validate_accuracy = [] 
r1_testing_accuracy = [] 
```

```python
for layer_type in r1_layer: 
    r1_model = build_model(encoder_width_1=512, pooling_method=layer_type, pooling_size=5)
    history = train_model(r1_model, epochs=5, verbose=1)
    
    # find accuracy
    r1_trained_X = r1_model.predict(X_train)
    r1_validated_X = r1_model.predict(X_validate)
    r1_tested_X = r1_model.predict(X_test)

    r1_binary_training_accuracy = tf.keras.metrics.BinaryAccuracy()
    r1_binary_training_accuracy.update_state(y_train, r1_trained_X)
    r1_train_accuracy.append(r1_binary_training_accuracy.result().numpy())

    r1_binary_validated_accuracy = tf.keras.metrics.BinaryAccuracy()
    r1_binary_validated_accuracy.update_state(y_validate, r1_validated_X)
    r1_validate_accuracy.append(r1_binary_validated_accuracy.result().numpy())

    r1_binary_testing_accuracy = tf.keras.metrics.BinaryAccuracy()
    r1_binary_testing_accuracy.update_state(y_test, r1_tested_X)
    r1_testing_accuracy.append(r1_binary_testing_accuracy.result().numpy())
    
```

# Part 3. Final Model 

**Based on results above, the best model has the parameters:**
* Pooling layer: avg / glb 
* Encoder width: 256
* Epoch size: 10 
* Activation type: sigmoid 

```python
# final model 
final_model = tf.keras.models.Sequential([
    mobile_net,
    tf.keras.layers.AvgPool2D(pool_size=(2, 2)),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=256, activation="relu"),
    tf.keras.layers.Dense(units=1, activation="sigmoid")
]) 
final_model.compile(
        optimizer='sgd',
        loss='binary_crossentropy',
        metrics=[tf.keras.metrics.BinaryAccuracy(name="binary_accuracy",threshold=0.5)]
)
batch_size = 64
final_model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=batch_size, 
    verbose=1, 
    validation_data=(X_validate, y_validate)
)
```

```python
final_trained_X = final_model.predict(X_train)
final_validated_X = final_model.predict(X_validate)
final_tested_X = final_model.predict(X_test)

final_binary_training_accuracy = tf.keras.metrics.BinaryAccuracy()
final_binary_training_accuracy.update_state(y_train, final_trained_X)
final_default_model_train_accuracy = final_binary_training_accuracy.result().numpy()

final_binary_validated_accuracy = tf.keras.metrics.BinaryAccuracy()
final_binary_validated_accuracy.update_state(y_validate, final_validated_X)
final_default_model_validate_accuracy = final_binary_validated_accuracy.result().numpy()

final_binary_testing_accuracy = tf.keras.metrics.BinaryAccuracy()
final_binary_testing_accuracy.update_state(y_test, final_tested_X)
final_default_model_test_accuracy = final_binary_testing_accuracy.result().numpy()
```

```python
print("Training Set Accuracy:", final_default_model_train_accuracy)
print("Validation Set Accuracy:", final_default_model_validate_accuracy)
print("Test Set Accuracy:", final_default_model_test_accuracy)
```
